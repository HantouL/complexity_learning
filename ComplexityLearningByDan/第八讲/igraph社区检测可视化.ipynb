```python
N = 100
g = ig.Graph.GRG(N,0.2)
ig.plot(g, bbox=(0, 0, 400, 400))
```

```python
par = ig.Graph.community_edge_betweenness(g, clusters=5)
```

```python
ig.plot(par, bbox=(0, 0, 400, 800))
```

```python
parti = ig.Graph.community_edge_betweenness(g, clusters=5).as_clustering()
```

```python
ig.plot(parti, bbox=(0, 0, 400, 400))
```

```python
par2 = ig.Graph.community_walktrap(g, steps=4)
parti2 = ig.Graph.community_edge_betweenness(g, clusters=5).as_clustering()
```

```python
ig.plot(par2, bbox=(0, 0, 400, 800))
```

```python
ig.plot(parti2, bbox=(0, 0, 400, 400))
```
